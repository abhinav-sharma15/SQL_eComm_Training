```python
import duckdb
from google.colab import drive

drive.mount('/content/drive')

DB_PATH = "/content/drive/MyDrive/ecomm.duckdb"
con = duckdb.connect(DB_PATH)
```

```python
con.execute("""
select * from orders;

""").fetchdf()
```

```python
con.execute("""
select * from orders limit 10;

""").fetchdf()
```

```python
con.execute("""
select * from orders
where status = 'paid';

""").fetchdf()
```

```python
con.execute("""
select * from orders
where status = 'paid'
limit 10;

""").fetchdf()
```

```python
con.execute("""
select * from orders
order by total_amount desc
limit 10;

""").fetchdf()
```

```python
con.execute("""
select * from orders
where status = 'paid'
order by total_amount desc
limit 10;

""").fetchdf()
```

```python
con.execute("""
select * from orders
where order_ts >= '2024-07-01'
order by order_ts;
""").fetchdf()
```

```python
con.execute("""
select o.order_id, c.customer_id, o.status, c.country
from orders o
join customers c on o.customer_id = c.customer_id
where o.status = 'paid'
and c.country = 'US'
order by o.order_id
limit 10;
""").fetchdf()
```
