```python
import duckdb
from google.colab import drive

drive.mount('/content/drive')

con = duckdb.connect("/content/drive/MyDrive/ecomm.duckdb")
```

```python
con.execute("SHOW TABLES").df()
```

```python

con.execute("DESCRIBE customers;").df()
```

```python
con.execute("DESCRIBE orders;").df()
```

```python
con.execute("""
SELECT *
FROM customers
LIMIT 10;
""").fetchdf()
```

```python
con.execute("""
SELECT COUNT(*) AS customer_count
FROM customers;
""").fetchdf()
```

```python
con.execute("""
SELECT COUNT(*) AS order_count
FROM orders;
""").fetchdf()
```

```python
con.execute("""
SELECT
  MIN(order_ts) AS first_order,
  MAX(order_ts) AS last_order
FROM orders;
""").fetchdf()
```

```python
con.execute("""
SELECT
  o.order_id,
  o.order_ts,
  c.country,
  o.total_amount
FROM orders o
JOIN customers c
  ON o.customer_id = c.customer_id
LIMIT 10;
""").fetchdf()
```
