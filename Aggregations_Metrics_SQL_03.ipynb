```python
import duckdb
from google.colab import drive

drive.mount('/content/drive')

DB_PATH = "/content/drive/MyDrive/ecomm.duckdb"
con = duckdb.connect(DB_PATH)
```

```python
con.execute("""
select count(*) as total_orders from orders;

""").fetchdf()
```

```python
con.execute("""
select sum(total_amount) as total_revenue from orders
where status = 'paid';

""").fetchdf()
```

```python
con.execute("""
select avg(total_amount) as avg_order_value from orders
where status = 'paid';

""").fetchdf()
```

```python
con.execute("""
select status, count(*) as num_orders from orders
group by status;
""").fetchdf()
```

```python
con.execute("""
select c.country, sum(o.total_amount) as total_revenue
from orders o
join customers c on c.customer_id=o.customer_id
where o.status = 'paid'
group by c.country
order by total_revenue desc;

""").fetchdf()
```

```python
con.execute("""
select c.country, count(o.status) as paid_orders, sum(o.total_amount) as revenue, avg(o.total_amount) as avg_order_value
from orders o
join customers c on c.customer_id=o.customer_id
where o.status = 'paid'
group by c.country
order by revenue desc;

""").fetchdf()
```

```python
con.execute("""
SELECT
  country,
  SUM(total_amount)
FROM orders o
JOIN customers c
  ON o.customer_id = c.customer_id
WHERE o.status = 'paid';
""").fetchdf()
```
